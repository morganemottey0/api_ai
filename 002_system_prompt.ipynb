```python
# %%
# Cellule 1 — Imports et setup
import anthropic

client = anthropic.Anthropic(api_key="API_KEY")
messages = []
```

```python
# %%
# Cellule 2 — Fonctions
def add_user_message(messages, user_input):
    messages.append({"role": "user", "content": user_input})

def add_assistant_message(messages, response):
    messages.append({"role": "assistant", "content": response})

def chat(messages, system=None):
    params = {
        "model": "claude-opus-4-5",
        "max_tokens": 1000,
        "messages": messages,
    }
    
    if system:
        params["system"] = system
    
    message = client.messages.create(**params)
    return message.content[0].text
```

```python
# %%
# Cellule 3 — Boucle chatbot
while True:
    user_input = input("You: ")
    if user_input.lower() == "quit":
        break
    
    add_user_message(messages, "Write a Python function that checks a string for duplicate characters and returns a list of the duplicates.")
    response = chat(messages, system="You are a Python engineer who writes very concise code")

    response
```
